```python
from urllib.request import urlopen
import spacy
import re
```

```python
input = urlopen("https://www.gutenberg.org/files/103/103-0.txt").read()
# print(input)
```

```python
input = input.decode("utf-8")
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
data = nlp(input)
```

```python
places = []

for token in data.ents:
  place = ""
  if(token.label_ == "LOC"):
    for text in nlp(token.text): 
      if text.pos_ != "DET":
        place = place + text.text  + " "

    if(place.lower().strip() not in places):
       places.append(place.lower().strip())

print(places)
```
